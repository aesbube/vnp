```python
data = pd.merge(left=data_feat, right=data_res, right_index=True, left_index=True)
```

```python
features = data.columns[:-1].tolist()
print(features)
```

```python
for feature in features:
    data[feature] = data[feature].interpolate(method="linear")
```

```python
data = data.groupby(pd.Grouper(freq='30min')).mean()
```

```python
lag = 5
```

```python
periods = range(lag, 0, -1)
```

```python
cols = data.columns.tolist()
```

```python
for lag in periods:
    for column in cols:
        data[f"{column}_{lag}"] = data[column].shift(lag)
```

```python
data
```

```python
features = cols[:-1]
features
```

```python
data = data.drop(columns=features, axis=1)
```

```python
data = data.dropna(axis=0)
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
```

```python
x, y = data.drop(columns='PowerConsumption', axis=1), data['PowerConsumption']
```

```python
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.20, shuffle=False)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
train_x = scaler.fit_transform(train_x)
test_x = scaler.transform(test_x)
```

```python
scaler = MinMaxScaler()
train_y = scaler.fit_transform(train_y.to_numpy().reshape(-1, 1))
```

```python
lag = 5
```

```python
train_x = train_x.reshape(train_x.shape[0], lag, (train_x.shape[1] // lag))
test_x = test_x.reshape(test_x.shape[0], lag, (test_x.shape[1] // lag))
```

```python
from keras.models import Sequential
from keras.layers import Input, LSTM, Dense
```

```python
model = Sequential([
    Input((train_x.shape[1], train_x.shape[2],)),
    LSTM(64, activation="relu", return_sequences=True),
    LSTM(32, activation="relu"),
    Dense(1, activation="linear")
])
```

```python
model.compile(
    loss="mean_squared_error",
    optimizer="adam",
    metrics=["mean_squared_error"],
)
```

```python
history = model.fit(train_x, train_y, validation_split=0.20, epochs=16, batch_size=64, shuffle=False)
```

```python
sns.lineplot(history.history["loss"], label="loss")
sns.lineplot(history.history["val_loss"], label="val_loss")
```

```python
predicted = model.predict(test_x)   
```

```python
predicted = scaler.inverse_transform(predicted)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
(mean_absolute_error(test_y, predicted), mean_squared_error(test_y, predicted), r2_score(test_y, predicted))
```
